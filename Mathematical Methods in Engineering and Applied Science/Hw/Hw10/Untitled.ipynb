```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f1(t,P):
    return np.array([t - t**2 / 2 + p for p in P])
def f1_2(X,T):
    return np.array([1/(1+ (X - t + t**2/ 2)**2) for t in T])
```

```python
t = np.linspace(-5,5,100)
p = np.linspace(-4,4,8)
figs = f1(t,p)
for fig in figs:
    plt.plot(fig,t,c='b')
    
plt.xlim([-2,2])
plt.ylim([-1,4])
plt.xlabel(r'$x$')
plt.ylabel(r'$t$')

plt.savefig("HW_10_1")
```

```python
X = np.linspace(-5,5,100)
T = np.arange(0,7)
figs = f1_2(X,T)

for fig in figs:
    plt.plot(X,fig)
    

plt.xlabel(r'$x$')
plt.ylabel(r'$u(x,t)$')
legend =[r"$t = {}$".format(t) for t in T]
plt.legend(legend)

plt.savefig("HW_10_2")
```

```python

X = np.linspace(-5,5,2)
P =np.arange(-5,5)
for p in P:
    plt.plot(X,X + p,c ='black')
plt.xlim([0,5])
plt.ylim([0,5])


plt.plot(X,X)
plt.xlabel(r"$x$")
plt.ylabel(r"$y$")
plt.savefig("Hw10_3")
```

```python
def f2(X,t):
    return[x * t + t-x if(x<t) else x*t for x in X]
```

```python
T = [0,0.4,0.8,1,2]
X = np.linspace(-10,10,200)
for t in T:
    plt.plot(X,f2(X,t))
plt.xlim([0,5])
plt.ylim([-0.2,4])
plt.legend([r"$t = {}$".format(t) for t in T])
plt.xlabel(r'$x$')
plt.ylabel(r'$u$')
plt.savefig("Hw10_4")
```
