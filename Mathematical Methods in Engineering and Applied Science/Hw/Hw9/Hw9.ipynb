```python
import matplotlib.pyplot as plt


mu = 1
def f2(Y):
    x, y = Y
    return [ y , x*(mu - x**2)]



def f3(Y):
    
    x, y = Y
    mu = 20
    Fx = -x
    if(x<=-1):
        Fx =x+2
    if(x >= 1):
        Fx = x -2
    return [ mu*(y - Fx) , -x/mu]


def f4(Y):
    x, y = Y
    mu = -0.005
    return [ -y+mu * x + x* y**2  , x + mu*y - x**2]

def f5(Y):
    x, y = Y
    mu =0.001
    return [ y  , x**2 - y - mu]


x_lim = 0.4
y_lim = 0.4
x = np.linspace(-x_lim, x_lim, 20)
y = np.linspace(-y_lim, y_lim, 20)

X, Y = np.meshgrid(x, y)

u, v = np.zeros(X.shape), np.zeros(Y.shape)

NI, NJ = X.shape

for i in range(NI):
    for j in range(NJ):
        x = X[i, j]
        y = Y[i, j]
        yprime = f5([x, y])
        yprime = yprime / np.linalg.norm(yprime)
        u[i,j] = yprime[0]
        v[i,j] = yprime[1]
     

Q = plt.quiver(X,Y, u, v, color='r')


x = np.linspace(-x_lim*2,x_lim*2,20)

#x1 =np.linspace(-1,1,20)
#plt.plot(-1*x1 , x1,c='b')

#x1 =np.linspace(-4,-1,20)
#plt.plot(x1 , x1+2,c='b')
#x1 =np.linspace(1,4,20)
#plt.plot(x1 , x1-2,c='b')
#plt.axvline(x=0, c='b')
#plt.axhline(y=0)

#plt.plot(x -1, x-1,c='b')

#plt.plot((-2-3**(1/2))*x,x,c='b')
plt.scatter(np.sqrt(0.001),0,c='black')
plt.scatter(-np.sqrt(0.001),0,c='black')
#plt.scatter(3/2,0,c='black')


plt.axvline(x=0)
plt.axhline(y=0)

plt.xlabel('$x$')
plt.ylabel('$y$')
plt.xlim([-x_lim, x_lim])
plt.ylim([-y_lim, y_lim])
#plt.grid()
plt.savefig("9_HW_plots")
```

```python
x, mu = sp.symbols('x mu')
```

```python
eq = x**4 - 2*x**3 +x**2 - mu*x + mu + mu**3
```

```python
ar = sp.solve(eq.subs(mu,0.21),x)
```

```python
sp.N(ar[1])
```

```python
for  mus in np.linspace(-0.5,1,40):
    ar = sp.solve(eq.subs(mu,mus),x)
    for item in ar:
        root = sp.N(item)
        if(sp.im(root) == 0):
            
            plt.scatter(mus,sp.re(root),marker='.',c='r')
            
            
            
for  mus in np.linspace(-0.14,0,40):
    ar = sp.solve(eq.subs(mu,mus),x)
    for item in ar:
        root = sp.N(item)
        
        
        #plt.scatter(mus,sp.re(root),c='b')
        if(sp.im(root) < 1e-20):
            #print(root)
            plt.scatter(mus,sp.re(root),marker='.',c='r')
            
for  mus in np.linspace(0,0.5,40):
    ar = sp.solve(eq.subs(mu,mus),x)
    for item in ar:
        root = sp.N(item)
        if(sp.im(root) == 0):
            
            plt.scatter(mus,sp.re(root),marker='.',c='r')
        
plt.axvline(-0.14)
plt.axvline(-0)
plt.axvline(0.21)
plt.xlabel('$$')


plt.xlabel(r'$\mu$')
plt.ylabel(r'$x$')
plt.savefig('Hw_9')
```
